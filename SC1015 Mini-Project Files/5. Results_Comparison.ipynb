### Results Comparison

We compare the TPR,FPR,TNR,FNR and Overall Accuracy of the models

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

#Import results data
TPRCat=pd.read_csv('TPR_Cat.csv')
TPRCat=TPRCat.drop(TPRCat.columns[[0]],axis=1)
TPRNum=pd.read_csv('TPR_Num.csv')
TPRNum=TPRNum.drop(TPRNum.columns[[0]],axis=1)
TPR=pd.merge(TPRCat,TPRNum,left_index=True, right_index=True)

FNRCat=pd.read_csv('FNR_Cat.csv')
FNRCat=FNRCat.drop(FNRCat.columns[[0]],axis=1)
FNRNum=pd.read_csv('FNR_Num.csv')
FNRNum=FNRNum.drop(FNRNum.columns[[0]],axis=1)
FNR=pd.merge(FNRCat,FNRNum,left_index=True, right_index=True)

TNRCat=pd.read_csv('TNR_Cat.csv')
TNRCat=TNRCat.drop(TNRCat.columns[[0]],axis=1)
TNRNum=pd.read_csv('TNR_Num.csv')
TNRNum=TNRNum.drop(TNRNum.columns[[0]],axis=1)
TNR=pd.merge(TNRCat,TNRNum,left_index=True, right_index=True)

FPRCat=pd.read_csv('FPR_Cat.csv')
FPRCat=FPRCat.drop(FPRCat.columns[[0]],axis=1)
FPRNum=pd.read_csv('FPR_Num.csv')
FPRNum=FPRNum.drop(FPRNum.columns[[0]],axis=1)
FPR=pd.merge(FPRCat,FPRNum,left_index=True, right_index=True)

ACCCat=pd.read_csv('ACC_Cat.csv')
ACCCat=ACCCat.drop(ACCCat.columns[[0]],axis=1)
ACCNum=pd.read_csv('ACC_Num.csv')
ACCNum=ACCNum.drop(ACCNum.columns[[0]],axis=1)
ACC=pd.merge(ACCCat,ACCNum,left_index=True, right_index=True)

```

```python
#Plot TPR,FNR,TNR,FPR,ACCURACY, Overall Accuracy bar graphs
f, axes = plt.subplots(5, figsize=(20, 16))
c = ['darkblue', 'blue', 'purple', 'orchid', 'grey','silver','Orange','brown','yellow','red']

TPR.plot(kind='bar',figsize=(20, 16),ax = axes[0],xlabel='Popularity Category',ylabel='TPR',title='True Positive Rates for Models',color=c)

FNR.plot(kind='bar',figsize=(20, 16),ax = axes[1],xlabel='Popularity Category',ylabel='FNR',title='False Negative Rates for Models',color=c)

TNR.plot(kind='bar',figsize=(20, 16),ax = axes[2],xlabel='Popularity Category',ylabel='TNR',title='True Negative Rates for Models',color=c)

FPR.plot(kind='bar',figsize=(20, 16),ax = axes[3],xlabel='Popularity Category',ylabel='FPR',title='False Positive Rates for Models',color=c)

ACC.plot(kind='bar',figsize=(20, 16),ax = axes[4],xlabel='Popularity Category',ylabel='Accuracy',title='Accuracy per category for Models',color=c)


f.tight_layout()
plt.show()
```
